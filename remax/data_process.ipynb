```python
import pandas as pd
df = pd.read_parquet('./gsm8k_chinese/data/train-00000-of-00001.parquet')
df.head()
```

```python
prompts = []
answers = []
for question, answer in zip(df['question_zh-cn'], df['answer_only']):
    SYSTEM = """
        按照如下格式回答问题：
        <think>
        你的思考过程
        </think>
        <answer>
        你的回答
        </answer>
        """
        
    
    prompt = [
            {
                "role": "system",
                "content": SYSTEM,
            },
            {
                "role": "user",
                "content": question,
            }
        ]
    
    prompts.append(prompt)
    answers.append(answer)
```

```python
df_new = pd.DataFrame()
df_new['prompt'] = prompts
df_new['answer'] = answers
df_new.head()
```

```python
df_new.to_parquet('./data_gsm8k/gsm8k_train.parquet')
```
