```python
from transformers import AutoTokenizer, AutoModelForCausalLM, AutoConfig
import torch
from pretrain import LLM, Config
```

```python
t = AutoTokenizer.from_pretrained('/home/user/wyf/train_qwen3_next_from_scratch/result_pretrain/checkpoint-5000')
AutoConfig.register("moe_model", Config)
AutoModelForCausalLM.register(Config, LLM)
```

```python

model = AutoModelForCausalLM.from_pretrained('/home/user/wyf/train_qwen3_next_from_scratch/result_pretrain/checkpoint-5000')
```

```python
input_data = [t.bos_token_id] + t.encode('你是一个')
print(input_data)
```

```python
for token in model.generate({"input_ids":torch.tensor(input_data).unsqueeze(0), "labels":None}, t.eos_token_id, 100, stream=False,temperature=0.0, top_k=1, repetition_penalty=1):
    print(t.decode(token[0]))
```

```python
model = AutoModelForCausalLM.from_pretrained('/home/user/wyf/train_qwen3_next_from_scratch/result_sft/checkpoint-4000')
```

```python
input_data = t.apply_chat_template([{'role':'user', 'content':'什么是人工智能？'}])
print(input_data)
```

```python
for token in model.generate({"input_ids":torch.tensor(input_data).unsqueeze(0), "labels":None}, t.eos_token_id, 500, stream=False,temperature=0.0, top_k=1, repetition_penalty=1):
    print(t.decode(token[0]))
```

```python
import json
import pandas as pd
with open('/home/user/wyf/train_qwen3_next_from_scratch/result_pretrain/checkpoint-5000/trainer_state.json' , 'r', encoding='utf-8') as f:
    data = json.load(f)
    log_history = data['log_history']
    steps = []
    losses = []
    for i in log_history[:-1]:
        step = i['step']
        loss = i['loss']
        steps.append(step)
        losses.append(loss)

import seaborn as sns
import plotly.express as px

# 创建一个DataFrame
df = pd.DataFrame({
    'Step': steps,
    'Loss': losses
})

# 使用Plotly绘制散点图
fig = px.line(df, x='Step', y='Loss', title='pretrain损失变化',
                 labels={'Step': 'step', 'Loss': 'loss'},
                 )  # 添加趋势线可选

fig.update_layout(width=600, height=400)
# 显示图表
fig.show()
```
